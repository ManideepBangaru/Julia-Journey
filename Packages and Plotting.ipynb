```julia
## Packages

To see all Julia Packages -  https://pkg.julialang.org/

# To install a package

import Pkg
Pkg.add("Example")

using Example

hello("It's me, Manideep !!!")

# Pkg.add("Colors")

using Colors

palette = distinguishable_colors(100)

rand(palette,3,3)

## Plotting

# Pkg.add("Plots")

using Plots

x = -3:0.1:3 # x is an array which takes the values between -3 to +3 with a step size of 0.1

sq(arr) = arr^2
y = sq.(x)

**_Now, Let's load the GR backend_**

gr(size=(250,200))

plot(x, y,label = "line")
scatter!(x, y, label = "points")
```
